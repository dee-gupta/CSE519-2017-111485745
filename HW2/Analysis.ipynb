```python
df3 = df3.dropna(axis=1, how='all')
```

 Dropping all columns where less than 1000 values are present, since replacing the rest with mean wont be meaningful.
 I chose 1000 which is ~ 1% of the 90K

```python
for col in list(df3):
    if df3[col].count() < 1000:
        df3.drop(col, axis = 1, inplace=True)
```

```python
df3.shape            #Left with 49 column
```

### Dropping categorical features sinnce we dont know how to deal with them yet and  dropping duplicate columns too.

```python
df3 = df3.drop(['airconditioningtypeid', 'buildingqualitytypeid','calculatedbathnbr','heatingorsystemtypeid',
                        'latitude','longitude', 'pooltypeid10', 'pooltypeid2', 'pooltypeid7', 'regionidzip',
                        'rawcensustractandblock','regionidcity', 'regionidcounty', 
                        'regionidneighborhood', 'censustractandblock'],axis=1)
```

```python
df3.shape
```

### dividing data into train data and test data
- 80% train data and 20% train data

```python
msk = np.random.rand(len(df)) < 0.8

train = df[msk]

test = df[~msk]
```

### Using statsmodel ols linear regression as our baseline model

```python
import statsmodels.api as sm
x = train[["bedroomcnt","calculatedfinishedsquarefeet"]]
y = train["logerror"]
model = sm.OLS(y, x).fit()
predictions = model.predict(x)
model.summary()
```

## How well/badly does it work?

- OLS stands for Ordinary Least Squares and the method “Least Squares” means that we’re trying to fit a regression line that would minimize the square of the distance from the regression line.

- R-squared value is 0.007 which tells us that the given model can explain 0.7% of log errors, which are predicted.

- The coefficient for an independent variable are .0005 and 0.00000537 for bedroomcnt and calculatedfinishedsquarefeet respectively, the significance of this coefficient is that if bedroomcnt variable increases by 1, the predicted value of log error increases by 0.0005.

- As we know that zillow model is already a good model and improving over this would certainly be hard, that's why R2 value is coming very low.

- do for choosen features this model works okay.


## Which variable is the most important one?

- In my opinion bedroom count is the most important one because it has correlation with log error. Fitting a linear model only on bedroom count gives 0.006 R-2 value that mean its correlated with house price.
- keeping Bed room count in the featues is surely making out model better.

### Support Vector Machine Regression

```python
from sklearn.svm import SVR
import numpy as np
from sklearn.metrics import accuracy_score
regressor = SVR(C=1.0, epsilon=0.1)
featuredf = train.filter(['bedroomcnt', 'finishedsquarefeet12'], axis=1)
dfy = train["logerror"]
regressor.fit(featuredf, dfy)
print "score:",regressor.score(featuredf,dfy)

fetureTest = test.filter(['bedroomcnt', 'finishedsquarefeet12'], axis=1)
predY = regressor.predict(fetureTest)
trueY = test.logerror

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(trueY, predY)
print"Mean Squared Error :",mse
```

### Decision Tree Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor

treeRegressor = DecisionTreeRegressor(random_state=0)
X = train.filter(['bedroomcnt', 'garagetotalsqft'], axis=1)
Y = train['logerror']
regressor = treeRegressor.fit(X,Y)
print "Score: ",treeRegressor.score(X,Y)

testX = test.filter(['bedroomcnt', 'garagetotalsqft'], axis=1)
Y = treeRegressor.predict(testX)
actualY = test.logerror

mse = mean_squared_error(actualY, Y)
print"Mean Squared Error :",mse                          
```

### K nearest neighbour model

```python
#K NEAREST NEIGHBOUR MODEL
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(train[['bedroomcnt','bathroomcnt','garagetotalsqft']], train[['logerror']])
predict = knn.predict(train[['bedroomcnt','bathroomcnt','garagetotalsqft']])

actual = train[['logerror']]
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(actual, predict)
print "Mean Squared Error :",mse
```

## Which of your models minimizes the squared error?

- As we can observe that we tried out three different models i.e. Support Vector Machine Regression, Decision Tree Regression and K nearest neighbor models and tried to fit the models on different features.

- In our case the Mean Squared Error score is minimum for K nearest neighbor model which is 0.0265873775184.

- Other model produces higher Mean Squared Error value, as we know that lower the MSE value better our model will be.
